```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import seaborn as sns
import math
import time
np.random.seed(int(time.time()))
```

```python
def show_hist_for_random(steps_count, set_len):
#     steps_count = 100000
#     set_len = 10000
    rands = [np.random.rand(set_len)*10 for i in range(steps_count)]
    # print(*rands, sep="\n")
    rands_sum = [np.sum(rands[i]) for i in range(len(rands))]
    plt.hist(rands_sum, bins=int(len(rands_sum)/100) if len(rands_sum) > 10 else 5)
    plt.show()
```

```python
show_hist_for_random(10, 10)
```

```python
show_hist_for_random(10000, 10000)
```
