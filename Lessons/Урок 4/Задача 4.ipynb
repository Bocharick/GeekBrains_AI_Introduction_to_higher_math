**Повторите расчеты, взяв другие значения n и k.**

```python
import numpy as np
import itertools
```

```python
count = 0
for p in itertools.product("01234", repeat=3):
    print("".join(p))
    count += 1
print("Total count:", count)
```

```python
count = 0
for p in itertools.permutations("01234", 5):
    print("".join(p))
    count += 1
print("Total count:", count)
```

```python
count = 0
for p in itertools.combinations("012345", 4):
    print("".join(p))
    count += 1
print("Total count:", count)
```
