```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

**Нарисуйте трехмерный график двух параллельных плоскостей.**

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z1 = 42*X + 33*Y
Z2 = 42*X + 33*Y + 421
ax.plot_wireframe(X, Y, Z1)
ax.plot_wireframe(X, Y, Z2)
show()
```

**Нарисуйте трехмерный график двух любых поверхностей второго порядка.**

```python
# Двухполостный гиперболоид
fig = figure()
ax = Axes3D(fig)
a = 14
b = 16
c = 2
X = np.arange(-35, 35, 2.5)
Y = np.arange(-35, 35, 2.5)
X, Y = np.meshgrid(X, Y)
Z1 = np.sqrt((X**2/a**2 + Y**2/b**2 + 1)*c**2)
Z2 = -np.sqrt((X**2/a**2 + Y**2/b**2 + 1)*c**2)
ax.plot_wireframe(X, Y, Z1)
ax.plot_wireframe(X, Y, Z2)
show()
```

```python
# Гиперболический параболоид
fig = figure()
ax = Axes3D(fig)
a = 4
b = 5
X = np.arange(-35, 35, 2.5)
Y = np.arange(-35, 35, 2.5)
X, Y = np.meshgrid(X, Y)
Z = X**2/a**2 - Y**2/b**2
ax.plot_wireframe(X, Y, Z)
show()
```
