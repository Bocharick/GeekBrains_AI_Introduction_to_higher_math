```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import seaborn as sns
import math
# Для установки одинакового масштаба осей X,Y следует добавить plt.gca().set_aspect('equal', adjustable='box')
```

```python
x = np.linspace(-8, 12, 201)
y = (x-2)**2 + 1
plt.plot(x, y)
plt.xlim(-15, 15)
plt.gca().set_aspect('equal', adjustable='box')
plt.ylim(-10, 10)
plt.xlim(-10, 10)
```

```python
s = 3*np.pi/2
x1 = x*np.cos(s)+y*np.sin(s)
y1 = -x*np.sin(s)+y*np.cos(s)
plt.plot(x1, y1)
plt.gca().set_aspect('equal', adjustable='box')
plt.ylim(-10, 10)
plt.xlim(-10, 10)
```

```python
s = np.pi/2
a11 = np.cos(s)
a12 = np.sin(s)
a21 = -np.sin(s)
a22 = np.cos(s)
x2 = x * a11 + y * a12
y2 = x * a21 + y * a22
plt.plot(x2, y2)
y3 = -x*a11/a12
y4 = -x*a21/a22
plt.plot(x, y3)
plt.plot(x, y4)
plt.ylim(-10, 10)
plt.xlim(-10, 10)
plt.gca().set_aspect('equal', adjustable='box')
```
