```python
from scipy.optimize import fsolve
```

**Решите систему уравнений:**  
y = x2 – 1  
exp(x) + x∙(1 – y) = 1  


```python
x = np.linspace(-10, 10, 301)
```

```python
y1 = x**2 - 1
y2 = (np.exp(x) + x - 1)/x
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.ylim(-3, 30)
plt.xlim(-6, 6)
plt.grid(True)
plt.gcf().set_size_inches(12.11, 10.24)
# plt.gca().set_aspect('equal', adjustable='box')
```

```python
def equations(p):
    x, y = p
    return (x**2-1-y, np.exp(x) + x * (1 - y) - 1)
```

```python
x1, y1 = fsolve(equations, (-2, 5))
print(x1, y1)

x1, y1 = fsolve(equations, (3, 7))
print(x1, y1)

x1, y1 = fsolve(equations, (5, 20))
print(x1, y1)

```

```python
def eq2(x):
    return ((x**2-1 - (np.exp(x) + x - 1)/x)**2)
```

```python
x_ = fsolve(eq2, (-2))
print(x_, x_**2-1)

x_ = fsolve(eq2, (3))
print(x_, x_**2-1)

x_ = fsolve(eq2, (5))
print(x_, x_**2-1)
```

**Решите систему уравнений и неравенств:**  
y = x2 – 1  
exp(x) + x∙(1 – y) > 1  

```python
x = np.linspace(-5, 5, 301)
x_below_zero = np.linspace(-5, 0-0.1, 301)
x_above_zero = np.linspace(0+0.1, 5, 301)

y1 = x**2 + 1
y2 = (np.exp(x_below_zero) + x_below_zero - 1)/x_below_zero
y3 = (np.exp(x_above_zero) + x_above_zero - 1)/x_above_zero


plt.plot(x, y1)
plt.plot(x_below_zero, y2)
plt.plot(x_above_zero, y3)
plt.xlim(-6, 6)
plt.ylim(0, 30)
plt.grid(True)
plt.gcf().set_size_inches(12.11, 10.24)
# plt.gca().set_aspect('equal', adjustable='box')
```
