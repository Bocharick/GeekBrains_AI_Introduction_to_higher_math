```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import seaborn as sns
```

```python
x = np.linspace(-25, 25, 1000)
```

```python
k1 = 1
k2 = 0.25
```

```python
y1 = np.cos(k1 * x)
y2 = np.cos(k2 * x)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(["y(%.2f*x)" % k1, "y(%.2f*x)" % k2], loc=4)
plt.show()
```
