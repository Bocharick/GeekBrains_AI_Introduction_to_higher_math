```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import seaborn as sns
import math
# Для установки одинакового масштаба осей X,Y следует добавить plt.gca().set_aspect('equal', adjustable='box')
```

```python
x = np.linspace(-25*np.pi, 25*np.pi, 2000)
```

```python
def y(k, a, b):
    return k*np.cos(x-a)+b
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

major_ticks = np.arange(-60, 60, 20)
minor_ticks = np.arange(-60, 60, 5)

ax.set_xticks(major_ticks)
ax.set_xticks(minor_ticks, minor=True)
ax.set_yticks(major_ticks)
ax.set_yticks(minor_ticks, minor=True)

ax.grid(which='both')

ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

plt.plot(x, y(10,2,3))
plt.plot(x, y(-10,-2,-3))
plt.plot(x, y(15,7,10))
plt.xlim(-12, 15)
plt.ylim(-20, 35)
plt.show()
```
