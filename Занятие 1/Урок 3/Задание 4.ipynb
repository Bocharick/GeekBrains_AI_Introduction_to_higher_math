**Решите систему уравнений:**

**Решите систему уравнений и неравенств:**  
y = x2 – 1  
exp(x) + x∙(1 – y) > 1  

```python
x = np.linspace(-5, 5, 301)
x_below_zero = np.linspace(-5, 0-0.1, 301)
x_above_zero = np.linspace(0+0.1, 5, 301)

y1 = x**2 + 1
y2 = (np.exp(x_below_zero) + x_below_zero - 1)/x_below_zero
y3 = (np.exp(x_above_zero) + x_above_zero - 1)/x_above_zero


plt.plot(x, y1)
plt.plot(x_below_zero, y2)
plt.plot(x_above_zero, y3)
plt.xlim(-6, 6)
plt.ylim(0, 40)
```
