**Напишите код, который будет переводить полярные координаты в декартовы.**  


```python
R = 5
alpha = 38 # угол в градусах
x = R * np.cos(alpha*np.pi/180)
y = R * np.sin(alpha*np.pi/180)
print("R = %.3f, Alpha = %.3f" % (R, alpha), "-->", "(%.3f, %.3f)" % (x, y))
```

**Напишите код, который будет рисовать график окружности в полярных координатах.**  


```python
x = np.linspace(0, 15, 100)
R = 3
y = x*0 + R
plt.polar(x, y)
plt.show()
```

**Напишите код, который будет рисовать график прямой линии в полярных координатах.**  


```python
ros = np.linspace(20, 80, 100)

phis = np.arccos(20/ros)+np.pi/6
phis2 = -np.arccos(20/ros)+np.pi/6

plt.polar(phis, ros)
plt.polar(phis2, ros)
plt.show()
```
