```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import seaborn as sns
import math
```

```python
# Окружность
# a,b - координаты цетнра окружности
a = 2397234
b = 141245

R = 10232

x = np.linspace(a-R, a+R, 1000)

y1 = []
y2 = []

for i in range(len(x)):
    y1_ = b + math.sqrt(abs(R**2 - (x[i] - a)**2))
    y1.append(y1_)
    y2_ = b - math.sqrt(abs(R**2 - (x[i] - a)**2))
    y2.append(y2_)
        

plt.gca().set_aspect('equal', adjustable='box')
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# Эллипс
a = 42
b = 33

x = np.linspace(-a, a, 10000)

y1 = []
y2 = []
for i in range(len(x)):
    y1_ = math.sqrt((1 - x[i]**2 / a**2)* b**2)
    y1.append(y1_)
    y2_ = -math.sqrt((1 - x[i]**2 / a**2)* b**2)
    y2.append(y2_)
    
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
# Гипербола
a = 3
b = 4

x1 = np.linspace(-a-10, -a, 100)
x2 = np.linspace(a, a+10, 100)
y1_1 = []
y1_2 = []
y2_1 = []
y2_2 = []
for i in range(len(x1)):
    y1_1_ = math.sqrt((x1[i]**2 / a**2 - 1)*b**2)
    y1_1.append(y1_1_)
    y1_2_ = -math.sqrt((x1[i]**2 / a**2 - 1)*b**2)
    y1_2.append(y1_2_)
    
    y2_1_ = math.sqrt((x2[i]**2 / a**2 - 1)*b**2)
    y2_1.append(y2_1_)
    y2_2_ = -math.sqrt((x2[i]**2 / a**2 - 1)*b**2)
    y2_2.append(y2_2_)
    
plt.gca().set_aspect('equal', adjustable='box')
plt.plot(x1, y1_1)
plt.plot(x1, y1_2)
plt.plot(x2, y2_1)
plt.plot(x2, y2_2)
plt.show()
```
